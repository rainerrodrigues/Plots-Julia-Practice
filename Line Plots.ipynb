```julia
using  Plots
```

```julia
x = range(0,10,length=100)
y =  sin.(x)
plot(x,y)
```

```julia
x = range(0,10,length=100)
y1 = sin.(x)
y2 = cos.(x)
plot(x,[y1 y2])
```

```julia
y3 = @. sin(x)^2 - 1/2 #equivalent to y3 = sin.(x).^2 .- 1/2
plot!(x,y3)
```

```julia
x = range(0,10,length=100)
y1= sin.(x)
y2= cos.(x)
p = plot(x,[y1 y2])

y3 = @. sin(x)^2 - 1/2
plot!(p,x,y3)
```

```julia
savefig("myplot.png")      # saves the CURRENT_PLOT as a .png
savefig(p, "myplot.pdf")   # saves the plot from p as a .pdf vector graphic
```

```julia
png("myplot")
Plots.pdf(p, "myplot")
```
